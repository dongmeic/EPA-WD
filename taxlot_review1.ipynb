```python
def create_ORTaxlots(x):
    lst = x.split(' ')
    if x in ['HWY 26', 'ISLAND', 'ROW', 'UPRR', 'VACATED', '']:
        txlot = ' '
    elif lst[-2] in ['U', 'L', 'A']:
        x = re.split('U|L|A', x)[0].strip()
        lst = x.split(' ')
    else:
        txlot = ' '
    if len(lst) == 5:
        trsqq = (lst[0] + lst[1] + lst[2] + lst[3]).rstrip("0")
    elif len(lst) == 4:
        trsqq = (lst[0] + lst[1] + lst[2]).rstrip("0")
    elif len(lst) == 3:
        trsqq = (lst[0] + lst[1]).rstrip("0")
    else:
        print(f'the list lengh is {len(lst)}, and the value is {x}')
    if len(lst) < 3:
        txlot = ' '
    else:
        txlot = create_ORTaxlot(cnt_code=33, trsqq=trsqq, lot=lst[-1])
    return txlot
```

```python
create_ORTaxlots(txgdf.TAXLOTS.values[0])
```

```python
txgdf.loc[~txgdf.TAXLOTS.isnull(), 'ORTaxlot'] = txgdf[~txgdf.TAXLOTS.isnull()].TAXLOTS.apply(lambda x: create_ORTaxlots(x))
```
